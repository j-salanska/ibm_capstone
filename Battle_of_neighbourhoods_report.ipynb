**Applied Data Science Capstone**

**This is Applied Data Science Capstone - The Battle of Neighbours**

**1. Introduction/Business Problem**

This report may be helpful for people who want to start a business (restaurant) in Toronto and trying to follow the trends.

**2. Data**

Our problem requires location data that describes hotels and restaurants around Toronto city center,
specifically, location of hotels and restaurants, number of restaurants around hotels, foot traffic in
restaurants and their category.
To retrieve this data, the Foursquare API was utilized, as it provides valuable location data that powers
some of the world famous applications. Therefore, two Major datasets were retrieved from Foursquare,
Hotels and Restaurants.

For the hotels dataset, a “Search” call was made to the Foursquare API with the search query “Hotel”.
The aim was to get the hotels around Toronto city center. However, after the call was made, results needed
to be organized into a Pandas data frame.
Further cleaning was made to keep only data attributes related to the name and location of each venue
as well as extracting the venues’ categories to clearly verify the results.
After results were organized and viewed in a Pandas data frame, it turned out that some results had
categories other than “Hotel”. Therefore, these results were dropped from the data frame as our focus is
on hotels only. 

**3. Methodology**

Firstly, the best touristing areas were found. Then the retreived hotels were clustered using K-means clustering.

The next step was to analyze the competition by finding the trendiing restaurants.

Then the evaluation was provided - simply counting the number of restaurants in each cluster to choose the one with the
lowest level of competition, and then count the number of hotels in that cluster and compare it to the
other clusters to evaluate opportunities. Simple value counting techniques were used in this evaluation.

**4. Results**

After grouping the hotels in three clusters by the K-Means Clustering algorithm, the resulting data frame
was visualized by assigning different color to each cluster.

The second output of the K-Means clustering algorithm was the location of the center in each cluster.

The KNN classification algorithm output was also visualized on a map to show each restaurant
in a different color based on the class to which it was assigned.

For the purpose of predicting completion levels, we counted the number of restaurants available in each
map.

Restaurants’ categories were inspected and we counted the frequency of each category to get an idea of
the most occurring categories and thus, the restaurants receiving the highest demand, since the data
frame used for this analysis already represents venues with the highest level of foot-traffic as explained
before.

**5. Discussion**

The analysis performed in this report was intended to tell us where to open the proposed restaurant and
what type of food it should supply.
After classifying the trending restaurants into the clusters generated by K-Means algorithm, and counting
the numbers in each, we found the area that has the lowest
level of competition.

Finally, we tried to learn what food category was in demand the most in Toronto city center by looking at
the top five frequent categories, and we found that four of them had an equal frequency of two
restaurants. However, by looking at these categories we find that among them are “Sushi Restaurant”,
“Japanese Restaurant”, and “Chinese Restaurant”. Since the latter two are also considered Asian food, we
can safely say that Asian food is the category in demand the most.
